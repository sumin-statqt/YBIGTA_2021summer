```python
import requests
from bs4 import BeautifulSoup
import re

actor_url = 'http://www.cine21.com/rank/person/content'

post_data_dict = dict()
post_data_dict['section'] = 'actor'
post_data_dict['period_start'] = '2020-07'
post_data_dict['gender'] = 'all'

actor_item_list = list()

for page in range(1,11):
    post_data_dict['page'] = page
    
    res = requests.post(actor_url, data=post_data_dict)

    soup = BeautifulSoup(res.text,'html.parser')

    actors = soup.select('li.people_li div.name')
    hits = soup.select('ul.num_info li strong')
    movies = soup.select('ul.mov_list')
    rankings = soup.select('li.people_li span.grade ')

    for index, actor in enumerate(actors):
        actor_item_dict = dict()
        actor_name = re.sub('\(\w*\)','',actor.text)
        actor_item_dict['이름'] = actor_name

        actor_hit = int(hits[index].text.replace(',',''))
        actor_item_dict['흥행지수'] = actor_hit

        movie_titles = movies[index].select('li a span')    
        movie_title_list = list()
        for movie_title in movie_titles:
            movie_title_list.append(movie_title.text)
        actor_item_dict['출연영화'] = movie_title_list

        actor_ranking = rankings[index].text
        actor_item_dict['랭킹'] = int(actor_ranking)


        actor_detail_full_url = 'http://www.cine21.com' + actor.select_one('a')['href']

        response_actor = requests.get(actor_detail_full_url)
        soup_actor = BeautifulSoup(response_actor.text,'html.parser')
        
        default_info = soup_actor.select_one('ul.default_info')
        actor_details = default_info.select('li')
    
        for actor_detail in actor_details:
            actor_detail_key = actor_detail.select_one('span.tit').text
            actor_detail_value = re.sub('<span.*>.*</span>', '', str(actor_detail))
            actor_detail_value = re.sub('<.*?>', '', actor_detail_value)
            actor_item_dict[actor_detail_key] = actor_detail_value

        actor_item_list.append(actor_item_dict)

print(actor_item_list)
```
